```python
import numpy as np
import matplotlib.pyplot as plt
from PIL.ImageColor import colormap

x = np.linspace(-10, 10, 21)
y = x**2 - 4*x + 4
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.plot(x, y)
plt.show()
```

```python
x = np.linspace(0, np.pi/2)
sin = np.sin(x)
cos = np.cos(x)
plt.plot(x, sin, linestyle='--', label='sin', c='red', marker='+')
plt.plot(x, cos, linestyle='-.', label='cos', c='blue', marker='*')
plt.xlabel('x')
plt.ylabel('y', rotation=0)
plt.legend()
plt.show()
```

```python
x = np.linspace(-5, 5, 11)
y = x**3

plt.subplot(2, 2, 1)
plt.plot(x, y, linestyle='--', label='x³', c='red')
plt.title('For f(x) = x³')
plt.legend()

x = np.linspace(0, 2 * np.pi, 100)
y = np.sin(x)

plt.subplot(2, 2, 2)
plt.plot(x, y, linestyle='-.', label='sin(x)', c='blue')
plt.title('For f(x) = sin(x)')
plt.legend()


x = np.linspace(-10, 10, 21)
y = np.exp(x)

plt.subplot(2, 2, 3)
plt.plot(x, y, linestyle='-', label='exp(x)', c='yellow')
plt.title('For f(x) = exp(x)')
plt.legend()


x = np.linspace(0, 10, 100)
y = np.log(x + 1)

plt.subplot(2, 2, 4)
plt.plot(x, y, linestyle=':', label='log(x)', c='Green')
plt.title('For f(x) = log(x)')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
x = np.random.uniform(0, 10, 100)
y = np.random.uniform(0, 10, 100)
colors = np.random.rand(100)
markers = np.random.choice(['o', 's', 'D', '^', 'v', 'p', '*'], 100)

for i in range(100):
    plt.scatter(x[i], y[i], c=[plt.cm.cool(colors[i])], s=100,
                alpha=0.7, marker=markers[i])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Scatter Plot of 100 Random Points")
plt.grid(True, linestyle='--', alpha=0.6)
plt.colorbar()
plt.show()
```

```python
x = np.random.normal(loc=0, scale=1, size=1000)

plt.hist(x, bins=30, color='green', edgecolor='red', alpha=0.5)

plt.xlabel("values")
plt.ylabel("frequency")
plt.title("Histogram of 1000 Random Points")
plt.grid(True, linestyle='--', alpha=0.6)

plt.show()
```

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)

Z = np.cos(X**2 + Y**2)

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('f(x, y)')
ax.set_title(r'$f(x, y) = \cos(x^2 + y^2)$')


surf = ax.plot_surface(X, Y, Z, cmap='plasma')
cbar = plt.colorbar(surf, ax=ax, shrink=0.6, aspect=10)
cbar.set_label('values')
plt.show()
```

```python
x = ['Product A', 'Product B', 'Product C', 'Product D', 'Product E']
y = [200, 150, 250, 175, 225]
colors = ['red', 'green', 'blue', 'yellow', 'purple']

plt.xlabel('Products')
plt.ylabel('Sales Values')
plt.title('Bar Chart of Products')

plt.bar(x, y, color=colors, alpha=0.8, edgecolor='brown', width=1)


plt.show()
```

```python
categories = ['T1', 'T2', 'T3', 'T4']
category_A = [1, 2, 3, 4]
category_B = [5, 6, 7, 8]
category_C = [9, 10, 11, 12]

fig, ax = plt.subplots()

ax.bar(categories, category_A, label='Category A', color='red')
ax.bar(categories, category_B, bottom=category_A, label='Category B', color='orange')
ax.bar(categories, category_C, bottom=np.array(category_A) + np.array(category_B), label='Category C', color='green')

ax.set_xlabel('Periods')
ax.set_ylabel('Values')
ax.set_title('Stacked Bar Chart')
ax.legend()

plt.show()
```
